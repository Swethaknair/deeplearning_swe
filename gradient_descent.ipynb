```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
X_b = np.c_[np.ones((100, 1)), X]
iterations = 1000
learning_rate = 0.01
stopping_threshold = 0.0001
theta = np.random.randn(2, 1)
def calculate_loss(X_b, y, theta):
    predictions = X_b.dot(theta)
    errors = predictions - y
    mse = np.mean(errors**2)
    return mse
def gradient_descent(X_b, y, theta, learning_rate, iterations, stopping_threshold):
    losses = []

    for i in range(iterations):
        predictions = X_b.dot(theta)
        errors = predictions - y
        gradients = 2/X_b.shape[0] * X_b.T.dot(errors)
        theta = theta - learning_rate * gradients
        loss = calculate_loss(X_b, y, theta)
        losses.append(loss)
        if i > 0 and abs(losses[i-1] - losses[i]) < stopping_threshold:
            break

    return theta, losses
theta, losses = gradient_descent(X_b, y, theta, learning_rate, iterations, stopping_threshold)
plt.scatter(X, y)
plt.plot(X, X_b.dot(theta), color='red', linewidth=3)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Gradient Descent Linear Regression')
plt.show()
plt.plot(losses)
plt.xlabel('Iterations')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Loss over Iterations')
plt.show()
print('Optimal parameters (theta):', theta)
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 2 + 3 * X + 1.5 * X**2 + np.random.randn(100, 1)
X_b = np.c_[np.ones((100, 1)), X, X**2]
iterations = 1000
learning_rate = 0.01
stopping_threshold = 0.0001
theta = np.random.randn(3, 1)
def calculate_loss(X_b, y, theta):
    predictions = X_b.dot(theta)
    errors = predictions - y
    mse = np.mean(errors**2)
    return mse
def gradient_descent(X_b, y, theta, learning_rate, iterations, stopping_threshold):
    losses = []

    for i in range(iterations):
        predictions = X_b.dot(theta)
        errors = predictions - y
        gradients = 2/X_b.shape[0] * X_b.T.dot(errors)
        theta = theta - learning_rate * gradients
        loss = calculate_loss(X_b, y, theta)
        losses.append(loss)

        if i > 0 and abs(losses[i-1] - losses[i]) < stopping_threshold:
            break

    return theta, losses
theta, losses = gradient_descent(X_b, y, theta, learning_rate, iterations, stopping_threshold)
plt.scatter(X, y)
X_range = np.linspace(0, 2, 100).reshape(-1, 1)
X_range_b = np.c_[np.ones((100, 1)), X_range, X_range**2]
plt.plot(X_range, X_range_b.dot(theta), color='red', linewidth=3)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Gradient Descent Quadratic Regression')
plt.show()
plt.plot(losses)
plt.xlabel('Iterations')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Loss over Iterations')
plt.show()
print('Optimal parameters (theta):', theta)
```
